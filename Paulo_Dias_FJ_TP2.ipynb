### Exercício 3 - Conversor de Moedas

##### Código corresponente

```python
try {
    while (true) {

        Map<Integer, Moeda> moedas = new HashMap<>();
        moedas.put(1, new Moeda(1, "Dólar Americano", "USD", 5.75));
        moedas.put(2, new Moeda(2, "Euro", "EUR", 6.05));
        moedas.put(3, new Moeda(3, "Libra Esterlina", "GBP", 7.29));


        System.out.print("Digite o montante em reais: R$ ");
        String currency = sc.nextLine();

        if (!numberValidator(currency)) {
            System.out.println("Valor inválido!");
            System.out.println(" ");
            continue;
        }

        double CCurrency = stringToDouble(currency);

        System.out.println(" ");
        System.out.println("Para qual moeda você deseja converter (Digite o número correspondente) ?");
        System.out.println("01 - USD");
        System.out.println("02 - EURO");
        System.out.println("03 - Libra Esterlina");
        String currencyConvertion = sc.nextLine();

        if (!numberValidator(currencyConvertion)) {
            System.out.println("Valor inválido - inicie o processo novamente!");
            System.out.println(" ");
            continue;
        }

        int CCurrencyConvertion = stringToInt(currencyConvertion);
        double converted;

        if (CCurrencyConvertion == 1) {
            converted = CCurrency / moedas.get(CCurrencyConvertion).getValue();
        } else if (CCurrencyConvertion == 2) {
            converted = CCurrency / moedas.get(CCurrencyConvertion).getValue();
        } else if (CCurrencyConvertion == 3) {
            converted = CCurrency / moedas.get(CCurrencyConvertion).getValue();
        } else {
            System.out.println("Valor inválido - inicie o processo novamente");
            System.out.println(" ");
            continue;
        }

        DecimalFormat df = new DecimalFormat("#.00");

        System.out.println(" ");
        System.out.println("O câmbio para " + moedas.get(CCurrencyConvertion).getName() +
                " está em " + moedas.get(CCurrencyConvertion).getValue() +
                ". Para R$ " + df.format(CCurrency) + " reais, o montante final é de " +
                "" + moedas.get(CCurrencyConvertion).getAcronym() + " " + df.format(converted));

        break;
    }
} catch (Exception ex) {
    System.out.println(String.format("Ocorreu um erro: %s", ex.getMessage()));
}
```

##### Observação: Este código foi criado com auxílio de uma classe Moedas:

```python
public class Moeda {
    private int id;
    private String name;
    private String acronym;
    private double value;

    public Moeda(int id, String name, String acronym, double value) {
        this.id = id;
        this.name = name;
        this.acronym = acronym;
        this.value = value;
    }

    public int getId() {
        return this.id;
    }

    public int setId() {
        return this.id = this.id;
    }

    public String getName() {
        return this.name;
    }

    public String setName() {
        return this.name = this.name;
    }

    public String getAcronym() {
        return this.acronym;
    }

    public String setAcronym() {
        return this.acronym = this.acronym;
    }

    public double getValue() {
        return this.value;
    }

    public double setValue() {
        return this.value = this.value;
    }
}
```

##### Este código inicia chamando o método Map<> para criação de uma lista com os dados das moedas, sendo eles: ID, name, acronym, value. Após inicializada o método HashMap<>(), a lista é populada com os dados das 3 moedas, USD, EUR e GBP. Após esse processo é solicitado ao usuário o montante em reais que ele deseja realizar a conversão, após o número ser validado através do método numberValidator(), o valor em reais é então convertido para o tipo double. Assim, é questionado para qual moeda o usuário deseja realizar a conversão, após validada a entrada, o valor é convertido para tipo int e também, é inicializada uma variável tipo double vazia, chamada converted, que tem por objetivo armazenar o valor final da conversão. O processo é filtrado através de if-else para identificar o cálculo correto que deve ser feito. Ao final é exibido em formato de formulário os valores em reais, o valor da moeda de conversão, e o valor convertido.

### Exercício 8 - Classificação de Triângulo

##### Código em execução:

```python
 try
{
    while (true)
    {
        System.out.print("Digite tres números espaçados para o lado A, B e C do triângulo  (Formato X X X): ");
        String sides = sc.nextLine();

        if (!numberValidatorTwo(sides))
        {
            System.out.println("Valor inválido!");
            System.out.println(" ");
            continue;
        }

        String[] numbers = sides.split(" ");
        double[] CNumbers = new double[numbers.length];

        for (int i = 0; i < numbers.length; i++){
            CNumbers[i] = Double.parseDouble(numbers[i]);
        }

        double sideA = CNumbers[0];
        double sideB = CNumbers[1];
        double sideC = CNumbers[2];

        if (sideA + sideB > sideC && sideA + sideC > sideB && sideB + sideC > sideA)
        {
            System.out.println("Os lados formam um triângulo válido.");
            System.out.println(" ");

            if (sideA == sideB && sideB == sideC)
            {
                System.out.println("Triângulo Equilátero: Todos os lados são iguais.");
                break;
            }
            else if (sideA == sideB || sideA == sideC || sideB == sideC)
            {
                System.out.println("Triângulo Isósceles: Dois lados são iguais.");
                break;
            }
            else
            {
                System.out.println("Triângulo Escaleno: Todos os lados são diferentes.");
                break;
            }
        }
        else
        {
            System.out.println("Os lados NÃO formam um triângulo válido, tente novamente.");
            System.out.println(" ");
            continue;
        }
    }
}
catch (Exception ex)
{
    System.out.println(String.format("Ocorreu um erro: %s", ex.getMessage()));
}
```

##### Neste codigo, similar ao que foi feito no exercício 2, são solicitados três valores ao usuário, com espaçamento entre eles. Após a validação através do método numberValidatorTwo(), os valores são convertidos para um array de doubles, via criação de uma String[] e split() para separar os dados na string validada e armazena-los no array, criado um array de doubles com o mesmo tamanho do array de strings anterimente criado, e um loop for para converter e popular o novo array double com o array de String. Os valores do array de doubles são então separados entre variáveis A, B e C e inseridos dentro de uma estrutura condicional if else, para identificar, se 1º formam um triângulo válido, e 2º qual tipo de triângulo formariam sendo válido.

### Exercício 10 - Jogo de Adivinhação

##### Código correspondente

```python
try
{
    int random = (int)(Math.random()* 101);

    int count = 5;
    while (count > 0)
    {
        System.out.println(String.format("Digite um chute, voce tem %s tentativas.", count));
        String guess = sc.nextLine();

        if (!numberValidator(guess))
        {
            System.out.println("Valor inválido!");
            System.out.println(" ");
            continue;
        }

        int CGuess = Integer.parseInt(guess);

        if(CGuess == random)
        {
            System.out.println(String.format("Acertou, o número misterioso era %s mesmo!.", CGuess));
            break;
        }
        else if (CGuess < random)
        {
            System.out.println("Seu chute foi abaixo, tente um pouco mais pra cima!");
            count--;
            continue;
        }
        else if (CGuess > random)
        {
            System.out.println("Seu chute foi acima, tente um pouco mais pra baixo!");
            count--;
            continue;
        }
    }
    if (count == 0)
    {
        System.out.println(String.format("Ops, suas tentativas se esgotaram :/, o número misterioso era %s.", random));
    }
}
catch (Exception ex)
{
    System.out.println(String.format("Ocorreu um erro: %s", ex.getMessage()));
}
```

##### Está estrutura se inicia com a criação de um número inteiro int aleatório entre 0 e 100. É inicializado um contador = 5 indicando a quantidade de tentativas para o acerto o usuário terá, será a nossa condicional para o loop while. Nesta etapa é solicitado ao usuário um chute, indicando quantas tentativas ele possui. Caso seja um chute muito acima, será indicado que ele deve chutar um pouco mais abaixo, caso seja abaixo será indicado que ele deve chutar um pouco mais acima, caso ele acerte o loop será encerrado e caso o contador cheque a 0, será encerrado com a mensagem dizendo que suas tentativas se esgotaram.

Após ambas validações ambos números são então convertidos para tipos int. ##### Esse exercício se inicia solicitando ao usuário uma frase aleatória com ao menos 5 palavras. Esse valor é armazenado num tipo string que após e transcrito para um string[] (Array) utilizando dos métodos trim() para aliminação de espaços no início e fim da frase, e do método split() para que ele separa várias palavras dentro da fraze em unidades dentro do array. Essa variável é contada com o método length() e o valor é exibido ao usuário com a quantidade de palavras que a frase contém.
